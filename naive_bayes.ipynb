# Naive Bayes


Modello di classificazione basato sul teorema di Bayes. Questa variante è chiamata "naive" (ingenua) per la sua assunzione di indipendenza tra le variabili. 

Utilizzando modelli di questa categoria, diamo dunque per scontato che le variabili predittive siano tutte indipendenti tra loro.


**Naive Bayes in Scikit-learn**



*   Caso studio: Iris Dataset





```python
from sklearn.datasets import load_iris
iris_data = load_iris()

iris_X = pd.DataFrame(iris_data["data"], columns=iris_data["feature_names"])
iris_y = pd.Series(iris_data["target_names"][iris_data["target"]])
```



*   Scikit-learn fornisce le implementaizoni di vari algoritmi Naive Bayes. Qui ne verranno analizzate alcune, testandole sul dataset Iris.





1.   Naive Bayes gaussiano

Questo modello assume una probabilità normale per la distribuzione delle variabili.



```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

X_train, X_test, y_train, y_test = train_test_split(iris_X, iris_y, test_size=1/3, random_state=1)
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
gaussian.score(X_test, y_test)
```



2. Naive Bayes multinomiale

Questo modello viene generalmente utilizzato per la classificazione di dati testuali ed è quindi probabile che non avrà buone perfomance sul dataset in questione.





```python
from sklearn.naive_bayes import MultinomialNB

multinomial = MultinomialNB()
multinomial.fit(X_train, y_train)
multinomial.score(X_test, y_test)
```



3. Naive Bayes di Bernoulli

In questo caso l'algoritmo lavora utilizzando la distribuzione di Bernoulli e si aspetta quindi delle variabili booleane. Essendo Iris un dataset multivariato, ovvero con molteplici variabili predittive, è probabile che otterrà scarsa precisione nella classificazione.



```python
from sklearn.naive_bayes import BernoulliNB

bernoulli = BernoulliNB()
bernoulli.fit(X_train, y_train)
bernoulli.score(X_test, y_test)
```
